```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
train_data = pd.read_csv("./src/data/test.csv", header=None).values
train_labels = train_data[:, 0]
train_data = train_data[:, 1:785]
train_data
```

```python
def plot_digits(features, labels, n_value):
    fig, axes = plt.subplots(1, 10, figsize=(10, 4))

    for i, ax in enumerate(axes.flat):
        if i < len(features):
            ax.imshow(features[i].reshape(28, 28), cmap="gray")
            ax.set_title(f"Digit {int(labels[i])}")
        plt.text(1.05, 0.5, f"Number of PCA component = {n_value}", size=12, ha="center", transform=fig.transFigure)
        ax.axis("off")
    plt.show()
```

```python
n_components_args = [10, 20, 30, 40, 50, 60, 70, 80]

for n in n_components_args:
    pca = PCA(n_components=n)
    reduced_train_data = pca.fit_transform(train_data)
    reduced_inverse_train_data = pca.inverse_transform(reduced_train_data[:10])
    plot_digits(reduced_inverse_train_data, train_labels[0:10], n)
    
```
